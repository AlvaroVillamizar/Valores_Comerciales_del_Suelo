```python
def Mun_dict():
    df = pd.read_csv('Divipola.csv')
    df['Departamento'] = df['Departamento'].str.lower()
    df['Municipio'] = df['Municipio'].str.lower()

    Departamentos = df['Departamento'].unique()
    Departamentos = np.append(Departamentos, ['bogota'])

    Municipios_col = {}

    for index, row in df.iterrows():
        departamento = row['Departamento']
        municipio = row['Municipio']

        if departamento in Municipios_col:
            Municipios_col[departamento].append(municipio)

        else:
            Municipios_col[departamento] = [municipio]
    return Municipios_col

Mun_dict = Mun_dict()
Mun_dict['bogota'] = ['bogota', 'bogota d.c', 'bogota, d.c', 'bogota, d.c.', 'bogota d.c.']
```

```python
def cleaning_municipio(df):
    test = pd.DataFrame(columns = ['municipio', 'departamento'])
    for i in range(df.shape[0]):
        if df.iloc[i,0] == '':   # Primer caso: municipios is Na
            if df.iloc[i,1] == '':
                df.loc[i] =  ['Sin dato', 'Sin dato']
            else:
                df.loc[i] =  ['Sin dato', df.iloc[i,1]]
        else:   
            mun_list = df.iloc[i,0].split(', ')
            if df.iloc[i,1] == '': # Departamento is Na
                if len(mun_list) == 1:
                    municipio = df.iloc[i,0].split(' ')
                    for mun in municipio:
                        for department in Mun_dict.keys():
                            if mun in Mun_dict[department]:
                                df.loc[i] =  [mun, department]
                else:
                    for item in mun_list: 
                        for department in Mun_dict.keys():
                            if item in Mun_dict[department]:
                                df.loc[i] =  [item, department]
            else:
                df.loc[i] =  [df.iloc[i,0], df.iloc[i,1]]
    return df
```

```python
import re

def extract_and_average(value):
    numbers = re.findall(r'\d+', str(value))
    if len(numbers) == 1:  
        return int(numbers[0])  
    elif len(numbers) > 1:  
        return (int(numbers[0]) + int(numbers[-1])) / 2
    else:
        return np.nan  
averages = np.array([extract_and_average(value) for value in chunk['antigüedad']])
```
